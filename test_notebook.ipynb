```python
seasons = ["Spring", "Summer", "Fall", "Cold"]
```

```python
print(seasons)
```

```python
for season in seasons:
    print(season)
```

```python
for season in seasons:
    print(season)
    if season =='Fall':
        break
```

```python
# use break prior to print to exit the loop, and Fall won't print
for season in seasons:
    if season == "Fall":
        break
    print(season)
```

```python
# use continue to pass through Fall and print the others

for season in seasons:
    if season == "Fall":
        continue
    print(season)
```

```python
# a set is an unordered collection of items.  Every set element is unique(no duplicates) and must be immutable
# the set itself is mutable, items can be added or removed from it

set = {'a','b', 'c'}
print(set)

#add method can be used to add an element to a set
```

```python
# you can initialize an empty set like this.  Careful!  It starts out as a dict
a = {}
print(type(a))
```

```python
# use update to add a list of items to a set
fruits = {"apple", "bananaa", "cherry"}
more_fruits = ["orange", "mango", "grapes"]
fruits.update(more_fruits)
print(fruits)
```

```python
# discard can also be used.  it's different than remove because the remove() method
# will raise an error if the specified item does not exist, discard will not

fruits.discard('cherry')
fruits.discard('vanilla')
```

```python
# get method can be used in dictionaries

seasonality = {
    "spring": "nice",
    "summer": "hot", 
    "fall": "colorful",
    "winter": "cold"
}

print(seasonality.get('summer'))
```

```python
seasonality = {
    "spring": "nice",
    "summer": "hot", 
    "fall": "colorful",
    "winter": "cold"
}

seasonality["winter"] = "cold af"
print(seasonality)
```

```python
# add a key value pair
seasonality["indian summer"] = "beautiful"
print(seasonality)
```

```python
#use pop to remove an item
seasonality.pop("indian summer")
```

```python
#use clear to remove all elements from the dictionary
seasonality.clear()
print(seasonality)
```
